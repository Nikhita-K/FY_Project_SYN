```python
import matplotlib.pyplot as plt
```

```python
# x-coordinates of left sides of bars 
left = ['Naive Bayes','Enc NB','KNN','Enc KNN', 'Random Forest','Enc RF', 'Decision Tree', 'Enc DT','Logistic Regression','Enc LR', 'SVM','Enc SVM']
```

```python
#accuracies of plain , enc
acc=[63,62,61,62,58,64,52,56,63,66,61,65]
```

```python
# labels for bars
label = ['Naive Bayes','Enc NB','KNN','Enc KNN', 'Random Forest','Enc RF', 'Decision Tree', 'Enc DT','Logistic Regression','Enc LR', 'SVM','Enc SVM']
```

```python
# plotting a bar chart
plt.figure(figsize=(16,7))
plt.bar(left, acc, label = label,
        width = 0.8, color=['lightpink','pink','purple','plum', 'paleturquoise','blue','aquamarine','grey','mistyrose','red','moccasin','yellow'])
# naming the x-axis
plt.xlabel('x - axis')
# naming the y-axis
plt.ylabel('y - axis')
# plot title
plt.title('Accuracy Comparison Plot--Airline DataSet')
  
# function to show the plot
plt.show()
```
